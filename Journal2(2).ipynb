```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,2*np.pi, 100)
```

```python
def equation_thing(a, b, x):
    line = a*np.exp(-b*x)

    return line
    
```

```python
b = (9/2*np.pi)*x+1

y_err = np.absolute(np.random.normal(loc=0.0, scale=2/b))

plt.errorbar(x,equation_thing(3,2,x), yerr=y_err,fmt="r^")
plt.xlabel("x")
plt.ylabel("3*e^-b*x")
```

```python
plt.savefig('awesomer_graph.png',bbox_inches='tight',dpi=400)
```
