```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,2*np.pi, 100)
```

```python
def equation_thing(a, b, x):
    line = a*np.exp(-b*x)

    return line
    
```

```python
plt.plot(x,equation_thing(3,2,x))
plt.xlabel("x")
plt.ylabel("3*e^-b*x")
```

```python
plt.savefig('awesome_graph.png',bbox_inches='tight',dpi=400)
```
